```python
x=df.values[:,[2,3]]
```

```python
x
```

```python
# using elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans

wsse= []
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=10)
    kmeans.fit(x)
    wsse.append(kmeans.inertia_) # within cluster sum off squared error
plt.plot(range(1,11),wsse)
plt.scatter(range(1,11),wsse)
plt.title('The elbow method')
plt.xlabel('number of cluster')
plt.ylabel('wsse')
plt.show()
```

```python
print(wsse)
```

```python
kmeans=KMeans(n_clusters=5, random_state=10)
y_pred=kmeans.fit_predict(x)

# kmeans.fit(x)--> training
# Y-pred=kmeans.predict(x)-->predicting
```

```python
y_pred
```

```python
kmeans.n_iter_ # iteration process # optional
```

```python
df['Clusters']=y_pred
df.head(10)
```

```python
sns.lmplot(data=df,x="Annual Income (k$)",y='Spending Score (1-100)',fit_reg=False,
          # no regression line
          hue='Clusters',palette='Set1')
#plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
# s = 300, c = 'yellow')
plt.show()
```

```python
kmeans.cluster_centers_
```

```python
sns.lmplot(data=df,x="Annual Income (k$)",y='Spending Score (1-100)',fit_reg=False,
          # no regression line
          hue='Clusters',palette='Set1')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
  s = 300, c = 'yellow')
plt.show()
```
